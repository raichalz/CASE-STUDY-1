# Case study on Presentation of data


1. Read the dataset to the python environment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data=pd.read_csv("cars_data.csv")
data.head()
```

2. Check for the null values present in the dataset.

```python
data.isnull().sum()
```

3. Plot a bar graph of male vs female buyers participated in the sales.

```python
d1=data['Buyer Gender'].value_counts()
print(d1)
```

```python
a1=list(d1.keys())
b1=list(d1.values)
plt.figure(figsize=(5,5))
plt.bar(a1,b1,color="maroon",width=.4)
plt.title("Male vs Female buyers participated in the sale")
plt.xlabel("Buyers")
plt.ylabel("Participation")
```

4. Find the top 5 cars based on their sales price.

```python
data.nlargest(5,'Sale Price')
```

5. Find the least 5 cars based on their Resell price.

```python
data.nsmallest(5,'Resell Price')
```
